## Partie 1 - Description du problème

Nous cherchons la ligue la plus prolifique en but à la saison 2015/2016.

## Partie 2 - Modèle

La base des matches européens pour la saison 2015 constitue notre modele.

## Partie 3 - Code

```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
data_Matche = pd.read_csv("projet/match.csv")
dm = data_Matche.dropna(subset =['home_player_1','home_player_2','home_player_3','home_player_4','home_player_5','home_player_6','home_player_7','home_player_8','home_player_9','home_player_10','home_player_11','away_player_1','away_player_2','away_player_3','away_player_4','away_player_5','away_player_6','away_player_7','away_player_8','away_player_9','away_player_10','away_player_11','away_team_goal','home_team_goal'])
dl = pd.read_csv('projet/league.csv')

def nb_buts(dataframe):
    if not((dataframe['home_team_goal'].values.size == 0 and  dataframe['away_team_goal'].values.size == 0)):
        return (dataframe['home_team_goal'] + dataframe['away_team_goal']).values[0]
    else:
        return 0
#print(nb_buts(dm.loc[dm['id'] == 1495]))
dico = dict()
dico['1'] = 0
dico['1729'] = 0
dico['4769'] = 0
dico['7809'] = 0
dico['10257'] = 0
dico['13274'] = 0
dico['15722'] = 0
dico['17642'] = 0
dico['19694'] = 0
dico['21518'] = 0
dico['19694'] = 0
dico['24558'] = 0
 #print('-------------')
nb_butes = 0
for i in dico.keys():
    vm = dm.loc[dm['league_id'] == int(i)]
    vmd = vm.loc[vm['season'] == '2015/2016']
    for j in vmd.index:
        #print(j)
        dico[i] += (nb_buts(vmd.loc[vmd['id'] == j]))
    nb_butes += dico[i]  
print(dico)
print("nombre de butes ",nb_butes)
```
